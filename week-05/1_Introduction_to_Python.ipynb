### List methods

```python
fruits = ["orange"]

fruits.append("lemon")
# fruits is now: ["orange", "lemon"]

fruits
```

```python
fruits = ["orange", "kiwi"]

fruits.insert(1, "lemon")
# fruits is now: ["orange", "lemon", "kiwi"]

fruits
```

```python
fruits = ["orange", "kiwi", "lime"]
fruits.pop()
# fruits is now: ["orange", "kiwi"]
fruits.pop(0)
# fruits is now: ["kiwi"]

fruits
```
