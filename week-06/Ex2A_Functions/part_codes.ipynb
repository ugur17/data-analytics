# Exercise 2.A Using built-in functions

## Lab 2. Parsing strings - part codes

```python
def get_supplier_code(part_code):
    # Find the colon and return everything before it.
    colon_index = part_code.find(':')
    return part_code[:colon_index]
```

```python
def get_product_num(part_code):
    # Find the colon and dash, then return what is between them.
    colon_index = part_code.find(':')
    dash_index = part_code.find('-')
    return part_code[colon_index + 1:dash_index]
```

```python
def get_size(part_code):
    # Find the dash and return everything after it.
    dash_index = part_code.find('-')
    return part_code[dash_index + 1:]
```

```python
part_code1 = 'ACME:123-L'
part_code2 = 'DI:12345-M'
part_code3 = 'ACE:1-12'
```

```python
supplier = get_supplier_code(part_code1)
part_number = get_product_num(part_code1)
size = get_size(part_code1)

print(f"For {part_code1}, the supplier is {supplier}. Part number {part_number} is available in size {size}.")
```

```python
supplier = get_supplier_code(part_code2)
part_number = get_product_num(part_code2)
size = get_size(part_code2)

print(f"For {part_code2}, the supplier is {supplier}. Part number {part_number} is available in size {size}.")
```

```python
supplier = get_supplier_code(part_code3)
part_number = get_product_num(part_code3)
size = get_size(part_code3)

print(f"For {part_code3}, the supplier is {supplier}. Part number {part_number} is available in size {size}.")
```

```python
part_codes = [part_code1, part_code2, part_code3]

for part_code in part_codes:
    supplier = get_supplier_code(part_code)
    part_number = get_product_num(part_code)
    size = get_size(part_code)
    print(f"For {part_code}, the supplier is {supplier}. Part number {part_number} is available in size {size}.")
```
