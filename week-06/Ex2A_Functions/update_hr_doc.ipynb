## Lab 3. Employee data

This lab exercise makes updates to information extracted from
an employee database.

```python
hr_list = [
    ('0123', 'HR', 'Rebecca Yang', '69000'),
    ('0121', 'IT', 'Mark Blick', '67000'),
    ('0068', 'IT', 'Kahna Larsen', '112000'),
    ('0234', 'OPS', 'Jim Smith', '54000')
]
```

```python
mark_record = list(hr_list[1])
mark_record[2] = 'Mark Blick-Hawley'
hr_list[1] = tuple(mark_record)
```

```python
print(hr_list)
```

```python
jim_record = list(hr_list[3])
jim_record[1] = 'CS'
jim_record[3] = '60000'
hr_list[3] = tuple(jim_record)
```

```python
print(hr_list)
```

```python
for employee_num, dept_code, name, salary in hr_list:
    print(f"{employee_num} | {dept_code} | {name} | {int(salary):,}")
```
