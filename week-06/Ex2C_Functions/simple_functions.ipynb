Lab 1 of Exercise 2.C, working with defining custom functions.

```python
def favorite_things(name, movie, music):
    return f"My name is {name}.\nMy favorite movie is {movie}.\nMy favorite musician or band is {music}."

print(favorite_things('Ceyhun Ugur', 'Kurtlar Vadisi', 'Manga'))
```

```python
def why_im_here(purpose):
    return purpose

print(why_im_here('I joined the Year Up United Data Analyst program to build job-ready data skills and start a stronger career path.'))
```

```python
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}."
```

```python
print(favorite_place('Istanbul', 'it has history, great food, and beautiful views'))
```

```python
print(favorite_place('Chicago', 'the city has great energy and amazing places to explore'))
```
